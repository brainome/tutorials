# 104 Using Brainome's Predictor CLI
The python predictor generated by Brainome has three modes.
1. Command line options
2. Validate test csv dataset
3. Classify unlabeled csv dataset
4. Advanced predictor usage

## Install brainome and generate a predictor
The predictor filename is predictor_104.py

The data sets are titanic_train.csv, titanic_validate.csv, and titanic_predict.csv

```python
# pip install brainome 
import sys
!{sys.executable} -m pip install --upgrade brainome --user --quiet
!brainome data/titanic_train.csv -y -f DT -split 90 -o predictor_104.py -modelonly -q
!head -50 predictor_104.py
```

## 1. Predictor --help
Predictors are really short and sweet. They just validate and classify data.

```python
!{sys.executable} predictor_104.py --help
```

## 2. -Validate a test csv data set
The predictor takes a csv data set identical to the training data set and compares outcomes.

```python
!{sys.executable} predictor_104.py -validate data/titanic_validate.csv
```

## 3. Classify a test data set.
The predictor can classify a similar csv data set sans target column. It will generate a "Prediction" column instead.

```python
!{sys.executable} predictor_104.py data/titanic_predict.csv > classifications_104.csv
print('Viewing classification predictions.')
import pandas as pd
classifications_output = pd.read_csv('classifications_104.csv')
classifications_output.head()
```
